```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor #LinearRegression공식, SGDRegressor(학습)
from sklearn.neural_network import MLPRegressor #학습(딥러닝)
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
carDF = pd.read_csv('data/cars.csv')
carDF.columns = ['speed','dist']
carDF
```

```python
x_data = carDF[['speed']]
```

```python
y_data = carDF['dist']
```

```python
#max_iter=1000 (학습횟수)
# alpha=0.0001 (런닝레이트)
# early_stopping=True( 코스트값이 변화가 없으면 학습종료)
modelSGD = SGDRegressor(max_iter=200,alpha=0.001,early_stopping=True,verbose=1)
modelSGD.fit( x_data, y_data)
```

```python
modelSGD.coef_
```

```python
modelSGD.intercept_
```

```python
y_predict = modelSGD.predict(x_data)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_predict,'r')
plt.show()
```
