```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cost( x,y,w):
    c =0 
    for i in range( len(x)):
        hx = w*x[i]
        c = c + (hx-y[i])**2
    return c/len(x)
```

```python
def gradient( x,y,w):  #합성함수의 미분구현
    c =0 
    for i in range( len(x)):
        hx = w*x[i]
        c = c + (hx-y[i])*x[i]
    return c/len(x)
```

```python
def show(x,y,w):
    hist = []
    for i in range( 200): #학습횟수
        c = cost( x,y, w)
        hist.append( c)
        print('i=', i, 'cost=', c, 'w=', w)
        g = gradient( x,y, w)
        w = w - 0.1*g  # learning rate * 기울기
    print( '최종:w', w )
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

```python
show( x_data, y_data, 10)
```
