```python
wine_df.corr()
```

```python
plt.figure(figsize=(25,12))
mask = np.zeros_like(wine_df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)]=True
cmap = sns.diverging_palette(10, 250, as_cmap=True)
sns.heatmap(wine_df.corr(), cmap=cmap, vmax=1, mask=mask, center=0, annot=True,
               fmt='.2f', square=True, linewidths=0.1, cbar_kws={"shrink": 1})
plt.xticks( rotation = 75)
plt.title("Diagonal Correlation Heatmap")
plt.show()
```

```python
x_data = wine_df.iloc[:,:-1]
y_data = wine_df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 7, stratify=y_data)
```

```python
model_logi = LogisticRegression( multi_class='multinomial')
model_logi.fit(x_train, y_train)
```

```python
print(model_logi.score(x_train, y_train))
print(model_logi.score(x_test, y_test))
```

```python
# confusion matrix
plt.figure(figsize=(25,12))
y_pred = model_logi.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='Pastel2', annot_kws={"size": 20})
plt.title('confusion matrix')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
