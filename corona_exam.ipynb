```python
df3 = pd.read_csv("/content/drive/MyDrive/YW_AI/study/corona/OBS_ASOS_TIM_20210112204135.csv", sep=",", encoding='cp949')
df3.head()
```

```python
df3.tail()
```

```python
df3.drop(['지점', '지점명'], axis=1, inplace=True)
df3['Date'] = pd.to_datetime(df3['일시'])
df3.head()
```

```python
df3['Month'] = df3['Date'].dt.month
df3['Day'] = df3['Date'].dt.day
df3.drop('일시', axis=1, inplace=True)
df3.head()
```

```python
df3['기온(°C)'].max()
```

```python
df3['기온(°C)'].min()
```

```python
df3['기온(°C)'].std()
```

```python
grouped_two = df3.groupby(['Month', 'Day'])

for key, group in grouped_two:
  print('* key :', key)
  print('* number :', len(group))
  print('* 평균기온 : %.2f °C' % group['기온(°C)'].mean())
  print('* 평균습도 : %.2f %%' % group['습도(%)'].mean())
  print('* 중간기온 : %.2f °C' % group['기온(°C)'].median())
  print('* 중간기온 : %.2f %%' % group['습도(%)'].median())
  print(group.head())
  print('\n')
```
